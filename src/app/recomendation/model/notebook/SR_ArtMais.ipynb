```python
teste = teste.convert_dtypes()
```

```python
teste
```

```python
sr_pivot = teste.pivot_table(columns='idusuario', index='idsubcategoria', values="visitas")
```

```python
sr_pivot
```

```python
sr_pivot.fillna(0, inplace=True)
```

```python
sr_pivot
```

```python
from scipy.sparse import csr_matrix
sr_sparse = csr_matrix(sr_pivot.astype(np.float))
```

#Criação do Modelo

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(sr_sparse)
```

#Fazendo recomendações

```python
distances, suggestions = model.kneighbors(sr_pivot.astype(np.float).iloc[0, :].values.reshape(1, -1))
```

```python
for i in range(len(suggestions)):
  print(sr_pivot.index[suggestions[i]])
```

```python
distances, suggestions = model.kneighbors(sr_pivot.astype(np.float).iloc[46, :].values.reshape(1, -1))
```
